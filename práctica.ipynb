```python
#mini proyecto del "Dia 1"


nom = input("Cual es tu nombre?: ")
edad = int(input("Cual es tu edad?: "))
peso = int(input("Cual es tu peso?: "))
altura = float(input("Cual es tu altura?: "))
IMC = peso / (altura ** 2)
IMC = round(IMC,2)


if IMC < 18.5 :
    clasificacion = "Bajo Peso"
elif IMC < 25:
    clasificacion = "Peso normal"
elif IMC < 30:
    clasificacion = "Sobrepeso"
elif IMC < 35:
    clasificacion = "Obesidad I"
elif IMC < 40:
    clasificacion = "Obesidad II"
else:
    clasificacion = "Obesidad III"

print("---------- PERFIL DEL USUARIO ----------")


print(f"Nombre: {nom}")
print(f"Edad: {edad}")
print(f"Peso: {peso}")
print(f"Altura: {altura}")
print(f"IMC: {IMC}")
print(f"Clasificación: {clasificacion}")

print("---------- PERFIL DEL USUARIO ----------")
```
