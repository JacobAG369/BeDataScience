```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
        "nombre": ["Ana", "Luis", "Carlos", "Jacob"],
        "edad":  [30, 33, 25, 20],
        "calificaciones": [90, 90, 76, 80]
}

df= pd.DataFrame(data)
df.describe()

plt.bar(df["nombre"], df["edad"])
plt.title("COMPARACIN DE EDADES")
plt.xlabel("ESTUDIANTE")
plt.ylabel("EDADES")
plt.grid(True)
plt.show()


plt.hist(df["edad"], bins=5)
plt.title("Distribucin de edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()


plt.plot(df["nombre"], df["edad"], marker="o", linestyle="--")
plt.title("Edades con estilo")
plt.grid(True)
plt.show()
```
